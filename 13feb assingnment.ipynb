```python
"""Write a python program to print Python Exception Hierarchy."""
#Ans=
import builtins

def print_exception_hierarchy():
    base_exceptions = [getattr(builtins, ex) for ex in dir(builtins) if isinstance(getattr(builtins, ex), type) and issubclass(getattr(builtins, ex), BaseException)]
    for exc in base_exceptions:
        print(f"{exc.__name__} (base class: {exc.__base__.__name__})")
        for subexc in exc.__subclasses__():
            print(f"  {subexc.__name__} (base class: {subexc.__base__.__name__})")
            
print_exception_hierarchy()
```
