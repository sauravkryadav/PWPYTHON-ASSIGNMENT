```python
#Q6

"""Ans:-
Solution:
Let's define the events:
A: Widget is defective
B: Widget tests positive for defects

We are interested in finding P(A|B), the probability that the widget is defective given a positive test result.

Given information:
P(A) = 0.05 (prior probability of a widget being defective)
P(B|A) = 0.98 (probability of a positive test result given that the widget is defective)
P(B|not A) = 0.05 (probability of a positive test result given that the widget is not defective)

We can calculate P(B), the probability of a positive test result, using the law of total probability:
P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
= 0.98 * 0.05 + 0.05 * 0.95
= 0.0995

Now we can calculate P(A|B) using Bayes' Theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)
= (0.98 * 0.05) / 0.0995
≈ 0.492

Therefore, the probability that the widget is actually defective, given a positive test result, is approximately 0.492 or 49.2%.

Bayes' Theorem allows us to update our prior knowledge (prior probability) about an event based on new evidence (conditional probabilities) to obtain a revised and more accurate probability (posterior probability).
In this example, even with a positive test result,
there is still a significant chance of a false positive due to the low prevalence of defective widgets. The revised probability indicates that there is a 49.2% chance that a widget is actually defective given a positive test result.
"""
```

```python
#Q7
"""Ans:-
To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5, we will use the following formula:

Confidence Interval = Sample Mean ± (Critical Value * Standard Error)

Where:

Sample Mean: 50
Standard Deviation: 5
Critical Value: This depends on the desired confidence level and the sample size. For a 95% confidence level with a large sample size (greater than 30), the critical value is approximately 1.96.
Standard Error: The standard error is calculated as the standard deviation divided by the square root of the sample size.
Now, let's calculate the confidence interval:

Sample Mean: 50
Standard Deviation: 5
Sample Size: Let's assume it is a large sample size of 100 for this example.

Standard Error = 5 / √100 = 0.5

Critical Value (for 95% confidence level): 1.96

Confidence Interval = 50 ± (1.96 * 0.5)
Confidence Interval = 50 ± 0.98
Confidence Interval = (49.02, 50.98)

Interpretation:
The 95% confidence interval for the sample of data with a mean of 50 and a standard deviation of 5 is (49.02, 50.98). This means that we are 95% confident that the true population mean falls within this interval"
"""
```

```python
#Q10

"""Ans:-

To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need to set up the null and alternative hypotheses, calculate the test statistic, and compare it to the critical value.

Null Hypothesis (H0): The weight loss drug is not significantly effective. The mean weight loss is not different from zero.
Alternative Hypothesis (Ha): The weight loss drug is significantly effective. The mean weight loss is different from zero.

Given information:
Sample size (n) = 50
Sample mean (x̄) = 6 pounds
Sample standard deviation (s) = 2.5 pounds

We will assume that the population follows a normal distribution. Since the population standard deviation is unknown, we will use a t-test.

Here are the steps to conduct the hypothesis test:

Set the significance level (α):
Let's use a 95% confidence level, which corresponds to a significance level (α) of 0.05.

Calculate the test statistic:
The test statistic for a one-sample t-test is calculated as follows:
t = (x̄ - μ) / (s / √n)

In this case:
x̄ = 6 pounds (sample mean)
μ = 0 pounds (null hypothesis, assuming no weight loss)
s = 2.5 pounds (sample standard deviation)
n = 50 (sample size)

t = (6 - 0) / (2.5 / √50)
t ≈ 17.89

Determine the critical value:
The critical value is obtained from the t-distribution table or statistical software, considering the degrees of freedom (df = n - 1). At a 95% confidence level with 49 degrees of freedom, the critical value is approximately ±2.009.

Compare the test statistic with the critical value:
If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

|17.89| > 2.009

Since |17.89| is greater than 2.009, we reject the null hypothesis."""
```

```python
#Q11
"""Ans:-
To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we will use the formula:

Confidence Interval = Sample Proportion ± (Critical Value * Standard Error)

Where:

Sample Proportion: The proportion of people in the sample who reported being satisfied with their job. In this case, it is 65%, which can be written as 0.65.
Critical Value: This depends on the desired confidence level. For a 95% confidence level, the critical value is approximately 1.96 for a large sample size.
Standard Error: The standard error is calculated as the square root of [(Sample Proportion * (1 - Sample Proportion)) / Sample Size].
Now, let's calculate the confidence interval:

Sample Proportion (p) = 0.65
Sample Size (n) = 500
Critical Value (for 95% confidence level) = 1.96

Standard Error = √[(p * (1 - p)) / n]
Standard Error = √[(0.65 * (1 - 0.65)) / 500]
Standard Error ≈ 0.0188

Confidence Interval = 0.65 ± (1.96 * 0.0188)
Confidence Interval = 0.65 ± 0.0369
Confidence Interval ≈ (0.6131, 0.6869)

Interpretation:
The 95% confidence interval for the true proportion of people who are satisfied with their job is approximately (0.6131, 0.6869).
This means that we are 95% confident that the true proportion of people in the population who are satisfied with their job lies within this interval."""
```
