```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
"""Ans:-
The Seaborn library is a popular data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five commonly used plots in Seaborn and their uses:

Scatter Plot:

Use: Display the relationship between two numeric variables.
Example: It can be used to explore the correlation between a person's age and their income.
Line Plot:

Use: Show the trend or pattern in data over a continuous interval.
Example: It can be used to visualize the stock prices of a company over a specific time period.
Bar Plot:

Use: Compare categorical variables by showing the frequency or value of each category.
Example: It can be used to compare the sales performance of different products in a store.
Histogram:

Use: Represent the distribution of a single numeric variable.
Example: It can be used to visualize the distribution of student exam scores in a class.
Box Plot:

Use: Display the summary statistics of a dataset, including median, quartiles, and outliers.
Example: It can be used to compare the distribution of house prices across different neighborhoods."""
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
#"timepoint" and y = "signal" for different events and regions.
"""Ans:-
"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Display the plot
plt.show()
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
#'pclass', y = 'age' and y = 'fare'.
#Note: pclass, age, and fare are columns in the titanic dataset.

"""Ans:-"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot for 'age'
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Passenger Class")
plt.show()

# Create the second box plot for 'fare'
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Passenger Class")
plt.show()
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
#parameter for the 'cut' column of the diamonds dataset.

"""Ans:-"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut' column
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")
plt.title("Histogram of Diamond Prices by Cut")
plt.show()
```

```python
#Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
"""Ans:-"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for 'species' column
sns.pairplot(data=iris_data, hue="species")
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
"""Ans:-"""
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Reshape the data into a pivot table
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create the heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")
plt.title("Flights Heatmap")
plt.show()
```
