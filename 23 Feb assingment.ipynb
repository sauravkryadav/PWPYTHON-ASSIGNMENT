```python
"""Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series?
Explain with an example.


Ans= DataFrame:-DataFrame is a two-dimensional data structure in Pandas that is used to represent tabular data in a flexible and easy-to-use way.
It is a collection of Pandas series objects that can be considerd as the sql tabels where the data are present in tabular format.
A DataFrame has both row and column labels and can be thought of as a dictionary of series objects, where each column is a Pandas Series.

Series:-Series is a one-dimensional labeled array in Pandas that can hold any data type.
It is a single column of a DataFrame and can be seen as a dictionary-like object where the keys are the index labels and the values are the data points."""

#code
import pandas as pd

data={"name":["Alice","Bob","Claire"],"Age":[25,30,27],"Gender":["Female","Male","Female"]}


#converting above data into pandas dataframe

var1=pd.DataFrame(data)
print("printing the data into Dataframe")
print()
print(var1)
print()
print("printing data in Series Format")
print()

#converting data into series format
var2=pd.Series(data)
print(var2)

```
