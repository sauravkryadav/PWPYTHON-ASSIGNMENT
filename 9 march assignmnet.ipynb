```python
#Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain withan example.

"""Ans:-
Probability Mass Function (PMF):
The Probability Mass Function (PMF) is a function that gives the probability of a discrete random variable taking on a specific value.
It assigns a probability to each possible outcome of the random variable.
For example, consider the tossing of a fair six-sided die. The random variable "X" represents the outcome of a single toss, which can take on values from 1 to 6.
The PMF of the die can be expressed as:

P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6

Here, the PMF provides the probabilities for each possible outcome of the die toss.

Probability Density Function (PDF):
The Probability Density Function (PDF) is a function that describes the probability distribution of a continuous random variable.
It provides the relative likelihood of observing a value within a given range.
For example, consider the height of adult males, which is a continuous random variable.
Let's assume the height follows a normal distribution with a mean of 175 cm and a standard deviation of 10 cm. The PDF of the height can be expressed as:

f(x) = (1 / (10 * sqrt(2 * pi))) * exp(-((x - 175) ^ 2) / (2 * 10^2))

In this example, the PDF provides the probability density of a male having a specific height "x" within the continuous range of possible values."""
```

```python
#Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

"""Ans:-The Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes on a value less than or equal to a specific value.
It provides information about the cumulative probability distribution of a random variable.

The CDF is denoted as F(x), where x is the value for which we want to calculate the cumulative probability.
The CDF is defined for both discrete and continuous random variables.

Example:-we can calculate the probability that a male is shorter than or equal to 180 cm using the CDF:

F(x <= 180) = ∫[175, 180] f(x) dx

Here, f(x) represents the Probability Density Function (PDF) of the height distribution.
The integral of the PDF from 175 to 180 gives the cumulative probability of a male being shorter than or equal to 180 cm"""
```

```python
#Q8: What is the z score? State the importance of the z score.

"""Ans:-The z-score, also known as the standard score, is a measure that quantifies how many standard deviations an individual data point or observation is away from the mean of a distribution.
It indicates the relative position of a data point within the distribution.

The importance of the z-score lies in several aspects:

Standardization: The z-score standardizes data, making it easier to compare and interpret values from different distributions.
By converting data into z-scores, we can assess how far a specific value deviates from the mean in terms of standard deviations.

Normal Distribution Analysis: The z-score is particularly useful for normal distributions. In a standard normal distribution (mean = 0, standard deviation = 1), the z-score represents the exact position of an observation in terms of its relative standing in the distribution.
It allows for easy identification of percentiles, probabilities, and comparisons within the standard normal distribution.

Outlier Detection: By calculating the z-score for individual data points, outliers can be identified as observations with extreme z-scores.
Typically, data points with z-scores greater than 2 or less than -2 are considered potential outliers.

Hypothesis Testing: The z-score is essential in hypothesis testing, specifically when comparing sample means to population means or when comparing two sample means.
The z-test, which uses the z-score, enables us to determine the statistical significance of observed differences.

Data Analysis and Inference: The z-score is widely used in various statistical analyses, such as constructing confidence intervals, estimating population parameters, and conducting parametric tests.
It simplifies calculations and allows for comparisons and inferences based on standard deviations."""
```

```python
#Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

"""Ans:-
The Central Limit Theorem (CLT) is a fundamental theorem in probability theory and statistics that states the following:

Given a large enough sample size from any population, the sampling distribution of the sample mean will be approximately normally distributed, regardless of the shape of the original population distribution.
Additionally, as the sample size increases, the approximation to the normal distribution improves.

The significance of the Central Limit Theorem lies in several aspects:

Normality Assumption: The CLT allows us to assume normality for the sampling distribution of the sample mean,
even if the population distribution is not normally distributed. This assumption is crucial for various statistical techniques and inference procedures that rely on the normal distribution, such as confidence intervals and hypothesis testing.

Approximation to Normal Distribution: The CLT enables us to approximate the distribution of the sample mean to a normal distribution, regardless of the shape of the population distribution.
This is particularly important because the normal distribution has well-known properties and simplifies statistical calculations.

Sample Size Independence: The CLT holds regardless of the underlying population distribution. This means that we can apply it to a wide range of populations, including those that are not well-defined or known.

Sample Mean Accuracy: The CLT ensures that as the sample size increases, the sample mean becomes a more accurate estimator of the population mean.
This property is vital for inferential statistics, as larger sample sizes lead to more reliable and precise estimates.

Statistical Inference: The CLT forms the foundation for many statistical inference procedures, such as constructing confidence intervals and conducting hypothesis tests.
It allows us to make probabilistic statements about population parameters based on sample data.

Real-World Applications: The CLT is widely applicable in various fields. It is utilized in areas such as survey sampling, quality control, medical research, economics, and social sciences.
It provides a robust framework for analyzing data and drawing reliable conclusions."""
```

```python
#Q10: State the assumptions of the Central Limit Theorem.

"""Ans:-The Central Limit Theorem (CLT) relies on certain assumptions in order to hold. The specific assumptions of the CLT are as follows:

Independent and Identically Distributed (IID) Sample: The observations in the sample must be independent of each other, meaning that the value of one observation does not affect the value of another.
Additionally, the observations must be drawn from the same underlying population and follow identical probability distributions.

Finite Mean and Variance: The population from which the sample is drawn must have a finite mean (μ) and finite variance (σ^2).
This assumption ensures that the mean and variance of the sample are well-defined.

Random Sampling: The sample should be selected randomly from the population. Random sampling ensures that the sample is representative of the population and avoids potential biases.

Sample Size: The CLT holds for large enough sample sizes. While there is no strict threshold for what constitutes a "large" sample size, as a general rule, a sample size of at least 30 is often considered sufficient for the CLT to provide a good approximation."""
```
